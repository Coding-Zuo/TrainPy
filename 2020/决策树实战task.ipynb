```python
import pandas as pd
import numpy as np
data=pd.read_csv('tree/iris_data.csv')
data.head()
```

```python
X=data.drop(['target','label'],axis=1)
y=data.loc[:,'label']
```

```python
from sklearn import tree
dc_tree=tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
dc_tree.fit(X,y)
```

```python
y_predict=dc_tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y,y_predict)
accuracy
```

```python
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength','SepalWidth','PetalLength','PetalWidth'],
               class_names=['setosa','versicolor','virginica'])
```

```python
dc_tree=tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=1)
dc_tree.fit(X,y)
fig=plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength','SepalWidth','PetalLength','PetalWidth'],
               class_names=['setosa','versicolor','virginica'])
```

```python
data=pd.read_csv('tree/anomaly_data.csv')
data.head()
```

```python
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'])
```

```python
x1=data.loc[:,'x1']
x2=data.loc[:,'x2']
fig2=plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(x1,bins=100)
plt.subplot(122)
plt.hist(x2,bins=100)
```

```python
x1_mean=x1.mean()
x1_sigma=x1.std()
x2_mean=x1.mean()
x2_sigma=x1.std()
```

```python
from scipy.stats import norm
x1_range=np.linspace(0,20,300)
x1_normal=norm.pdf(x1_range,x1_mean,x1_sigma)
x2_range=np.linspace(0,20,300)
x2_normal=norm.pdf(x2_range,x2_mean,x2_sigma)
```

```python
fig=plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(x1_range,x1_normal)
plt.subplot(122)
plt.plot(x2_range,x2_normal)
```

```python
from sklearn.covariance import EllipticEnvelope
ad_model=EllipticEnvelope()
ad_model.fit(data)
```

```python
y_predict=ad_model.predict(data)
pd.value_counts(y_predict)
```

```python
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',
            facecolor='none',edgecolor='red',s=150)
```

```python
ad_model=EllipticEnvelope(contamination=0.02)
ad_model.fit(data)
y_predict=ad_model.predict(data)
pd.value_counts(y_predict)
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',
            facecolor='none',edgecolor='red',s=150)
```
